```python
autos['odometer'] = autos['odometer'].str.replace('km', '')
autos['odometer'] = autos['odometer'].str.replace(',', '').astype(float)
autos.rename({'odometer': 'odometer_km'}, inplace=True)
```

### Exploring Price by Brand

```python
brands = autos['brand'].value_counts(normalize=True).head(20)
brand_mean_prices = {}
    
for i in range(len(brands)):
    price = autos[autos['brand'] == brands.index[i]]['price'].mean()
    brand_mean_prices[brands.index[i]] = price
    
print(brand_mean_prices)

sorted_dict = {}
sorted_keys = sorted(brand_mean_prices, key=brand_mean_prices.get)

for entry in sorted_keys:
    sorted_dict[entry] = brand_mean_prices[entry]

print('\n')

for key,val in sorted_dict.items():
    print('brand: {}, price: {:,.2f}'.format(key,val))
```

### Storing Aggregate Data in a DataFrame

```python
brands = autos['brand'].value_counts(normalize=True).head(20)
brand_mean_mileage = {}
    
for i in range(len(brands)):
    mileage = autos[autos['brand'] == brands.index[i]]['odometer_km'].mean()
    brand_mean_mileage[brands.index[i]] = mileage
    
print(brand_mean_mileage)

sorted_dict = {}
sorted_keys = sorted(brand_mean_mileage, key=brand_mean_prices.get)

for entry in sorted_keys:
    sorted_dict[entry] = brand_mean_mileage[entry]

print('\n')

for key,val in sorted_dict.items():
    print('brand: {}, mileage: {:,.2f}'.format(key,val))
```

```python
bpm_series = pd.Series(brand_mean_prices)
bom_series = pd.Series(brand_mean_mileage)
# print(bpm_series)

df = pd.DataFrame(bpm_series, columns=['mean_price'])
df['mean_mileage'] = bom_series
df
```
